```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from PIL import Image, ImageDraw

pil_base = Image.open('img/base.png')
(width, height) = pil_base.size
print('Width: %d' % width)
print('Height: %d' % height)
display(pil_base)
```

```python
dpi = 100
figsize = (width/dpi, height/dpi)
# fig, ax = plt.subplots(figsize=figsize, dpi=dpi)
# fig, ax = plt.subplots()
fig = plt.figure(figsize=figsize, dpi=dpi)
image = fig.figimage(pil_base, xo=0, yo=0, origin='upper')
# plt.axis('image')
ax = fig.gca()

scalebar = AnchoredSizeBar(
    transform = ax.transData,
    size = 300, 
    label = '1 cm', 
    loc = 'lower center',
    # pad=0.2,
    # border_pad=0.5,
    # sep=5
    # color='black',
    # frameon=False,
    size_vertical=0,
    # fontproperties=fontprops
)

ax.add_artist(scalebar)

ax.set_yticks([])
ax.set_yticklabels([])
ax.set_xticks([])
ax.set_xticklabels([])

# print('fig')
# display(fig)

# plt.imshow(pil_base)
```
